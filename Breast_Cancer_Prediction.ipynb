```python
raw_df['diagnosis']=raw_df['diagnosis'].map({'M':1,'B':0})
```

```python
from sklearn.model_selection import train_test_split
train_df,test_df=train_test_split(raw_df,test_size=0.2,random_state=42)
```

```python
print(f"Train Data Shape: {train_df.shape}")
print(f"Test Data Shape: {test_df.shape}")
```

```python
input_cols=list(train_df.columns)
input_cols.remove('diagnosis')
input_cols
```

```python
target_col='diagnosis'
```

```python
train_inputs=train_df[input_cols]
train_targets=train_df[target_col]
test_input=test_df[input_cols]
test_targets=test_df[target_col]
```

```python
test_targets
```

```python
train_inputs
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(raw_df[input_cols])
```

```python
train_inputs=scaler.transform(train_inputs)
test_input=scaler.transform(test_input)
```

```python
train_inputs
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(train_inputs,train_targets)
```

```python
print(input_cols)
```

```python
print(model.coef_.tolist())
```

```python
X_train_pred=model.predict(train_inputs)
X_test_pred=model.predict(test_input)
```

```python
from sklearn.metrics import accuracy_score
train_acc=accuracy_score(train_targets,X_train_pred)
test_acc=accuracy_score(test_targets,X_test_pred)
print(f"Train Accuracy: {train_acc}")
print(f"Test Accuracy: {test_acc}")
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(test_targets,X_test_pred)
print(cm)
```

```python
def predict_input(new_input):
  new_input=pd.DataFrame([new_input])
  new_input=scaler.transform(new_input)
  prediction=model.predict(new_input)
  if prediction[0]==0:
    print("The Breast Cancer is Benign")
  else:
    print("The Breast Cancer is Malignant")
```
